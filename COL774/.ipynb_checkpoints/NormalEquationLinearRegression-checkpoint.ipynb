```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
inputX = pd.read_csv('q3x.dat',sep="  ",header=None, engine = 'python')
X = inputX.as_matrix()
```

```python
X = np.insert(X,0,1,axis=1)
X = np.asmatrix(X)
```

```python
inputY = pd.read_csv('q3y.dat',sep="  ",header=None, engine = 'python')
Y = inputY.as_matrix()
Y = np.asmatrix(Y)
```

```python
XTX = np.transpose(X)*X
inv_XTX = np.linalg.inv(XTX)
XTY = np.transpose(X)*Y
theta = inv_XTX*XTY
```

```python
theta
```

```python
plt.plot(X[:,1],Y,'bo')
y = lambda x: theta[0,0]+theta[1,0]*x
plt.plot(X[:,1],y(X[:,1]))
plt.show()
```

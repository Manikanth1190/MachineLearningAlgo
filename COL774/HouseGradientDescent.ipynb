```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#read input
inputX = pd.read_table('q1x.dat',delimiter='\t',header=-1)
X = inputX.as_matrix()
```

```python
#read labels
inputY = pd.read_table('q1y.dat',header = -1)
Y = inputY.as_matrix()
```

```python
#plot data
plt.plot(X,Y,'bo')
plt.show()
```

```python
#define learning rate and precision
learningRate = 0.0001 #0.1 0.01 0.001 0.0001
precision = 0.001
```

```python
theta0 = 0
theta1 = 0
fx = lambda x: theta0+theta1*x
def ErrorSum(fx,X,Y):
    total = 0
    total1 = 0
    for i in range(0,len(X)):
        total = total + Y[i]-fx(X[i])
        total1 = total1 + X[i]*(Y[i]-fx(X[i]))
    return total[0],total1[0]
```

```python
while True:
    THETAPREV = np.matrix([[theta0],[theta1]])
    t0, t1 = ErrorSum(fx,X,Y)
    theta0=theta0+learningRate*t0
    theta1=theta1+learningRate*t1
    THETA = np.matrix([[theta0],[theta1]])
    #check convergence
    if(np.linalg.norm(THETA - THETAPREV)<precision):
        break
```
